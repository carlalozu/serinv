## 1. Block tridiagonal matrix

```python
import numpy as np
from scipy.linalg import lu_factor

# Create a sample matrix
A = np.array([[4, 3], [6, 3]])

# Perform LU decomposition with overwrite_a=True
lu, piv = lu_factor(A, overwrite_a=True)

# The original matrix A has been overwritten
print("Original matrix A after LU decomposition with overwrite_a=True:\n", A)

print(lu)

# Create a new sample matrix
B = np.array([[4, 3], [6, 3]])

# Perform LU decomposition with overwrite_a=False (default behavior)
lu, piv = lu_factor(B, overwrite_a=False)

# The original matrix B remains unchanged
print("\nOriginal matrix B after LU decomposition with overwrite_a=False:\n", B)

A.shape[0]
```

```python
%matplotlib widget

from sdr.lu.lu_decompose import lu_dcmp_tridiag
from sdr.lu.lu_solve import lu_slv_tridiag

nblocks = 5
blocksize = 2
symmetric = True
diagonal_dominant = True
seed = 63

A = matrix_generation.generate_blocktridiag(
    nblocks, blocksize, symmetric, diagonal_dominant, seed
)

# P_ref, L_ref, U_ref = la.lu(A)
lu_ref, p_ref = la.lu_factor(A)
L_sdr, U_sdr = lu_dcmp_tridiag(A, blocksize)

n_rhs = 3
B = np.random.randn(A.shape[0], n_rhs)


# --- Solving ---

X_ref = la.lu_solve((lu_ref, p_ref), B)
# Is equivalent to..
# Y_ref = la.solve_triangular(L_ref, B, lower=True)
# X_ref = la.solve_triangular(U_ref, Y_ref, lower=False)

fig, ax = plt.subplots(1, 3)
fig.suptitle("LU selected solver for block tridiagonal matrices")
ax[0].set_title("X_ref: Scipy lu solver")
ax[0].matshow(X_ref)

X_sdr = lu_slv_tridiag(L_sdr, U_sdr, B, blocksize)
ax[1].set_title("X_sdr: Selected lu solver")
ax[1].matshow(X_sdr)

X_diff = X_ref - X_sdr
ax[2].set_title("X_diff: Difference between X_ref and X_sdr")
ax[2].matshow(X_diff)
fig.colorbar(ax[2].matshow(X_diff), ax=ax[2], label="Relative error", shrink=0.4)


```

## 2. Block tridiagonal arrowhead matrix

```python
from sdr.lu.lu_decompose import lu_dcmp_tridiag_arrowhead
from sdr.lu.lu_solve import lu_slv_tridiag_arrowhead

nblocks = 5
diag_blocksize = 3
arrow_blocksize = 2
symmetric = True
diagonal_dominant = True
seed = 63

A = matrix_generation.generate_blocktridiag_arrowhead(
    nblocks, diag_blocksize, arrow_blocksize, symmetric, diagonal_dominant, 
    seed
)

# --- Decomposition ---
# P_ref, L_ref, U_ref = la.lu(A)
lu_ref, p_ref = la.lu_factor(A)
L_sdr, U_sdr = lu_dcmp_tridiag_arrowhead(A, diag_blocksize, arrow_blocksize)

n_rhs = 5
B = np.random.randn(A.shape[0], n_rhs)


# --- Solving ---
X_ref = la.lu_solve((lu_ref, p_ref), B)
# Is equivalent to..
# Y_ref = la.solve_triangular(L_ref, B, lower=True)
# X_ref = la.solve_triangular(U_ref, Y_ref, lower=False)

fig, ax = plt.subplots(1, 3)
fig.suptitle("LU selected solver for block tridiagonal arrowhead matrices")
ax[0].set_title("X_ref: Scipy lu solver")
ax[0].matshow(X_ref)

X_sdr = lu_slv_tridiag_arrowhead(L_sdr, U_sdr, B, diag_blocksize, arrow_blocksize)
ax[1].set_title("X_sdr: Selected lu solver")
ax[1].matshow(X_sdr)

X_diff = X_ref - X_sdr
ax[2].set_title("X_diff: Difference between X_ref and X_sdr")
ax[2].matshow(X_diff)
fig.colorbar(ax[2].matshow(X_diff), ax=ax[2], label="Relative error", shrink=0.4)
```

## 3. Block n-diagonals matrix

```python
from sdr.lu.lu_decompose import lu_dcmp_ndiags
from sdr.lu.lu_solve import lu_slv_ndiags

nblocks = 6
ndiags = 7
blocksize = 2
symmetric = False
diagonal_dominant = True
seed = 63

A = matrix_generation.generate_block_ndiags(
    nblocks, ndiags, blocksize, symmetric, diagonal_dominant, seed
)

# --- Decomposition ---
P_ref, L_ref, U_ref = la.lu(A)
# lu_ref, p_ref = la.lu_factor(A)
L_sdr, U_sdr = lu_dcmp_ndiags(A, ndiags, blocksize)

## call solve now
n_rhs = 1
B = np.random.randn(A.shape[0], n_rhs)

Y_ref = la.solve_triangular(L_ref, B, lower=True)
X_ref = la.solve_triangular(U_ref, Y_ref, lower=False)
# X_ref = la.lu_solve((lu_ref, p_ref), B)

X_sdr = lu_slv_ndiags(L_sdr, U_sdr, B, ndiags, blocksize)
#print("norm(X_ref - X_sdr) = ", np.linalg.norm(X_ref - X_sdr))

fig, ax = plt.subplots(1, 3)
ax[0].set_title("X_ref: Reference lu solver")
ax[0].matshow(X_ref)

ax[1].set_title("X_sdr: Selected lu solver")
ax[1].matshow(X_sdr)

X_diff = X_ref - X_sdr
ax[2].set_title("X_diff: Difference between X_ref and X_sdr")
ax[2].matshow(X_diff)
fig.colorbar(ax[2].matshow(X_diff), ax=ax[2], label="Relative error", shrink=0.4)

plt.show()
```
